Univariate analysis And Bivarient Analysis

Univariate analysis using Describe() method

```python
df.describe()
```

Joint Plot -used for plotting of numeric values

```python
sns.jointplot(x="Item_Outlet_Sales",y="Item_Weight",data=df,kind="reg")
```

```python
sns.jointplot(x="Item_Outlet_Sales",y="Item_Weight",data=df,kind="hex")
```

Bivarient Analysis using Scatterplot

```python
sns.scatterplot(x="Item_Outlet_Sales",y="Item_Weight",data=df)
```

```python
sns.pairplot(df) #Analysing the graph w.r.t numeric columns.
```

```python
sns.pairplot(df,hue="Item_Fat_Content")
```

Univariate analysis of Numeric Column

```python
sns.distplot(df['Item_Outlet_Sales'],kde=False) # inorder to view the frequency of occurence instead of probability distribution, put kde=False.
```

```python
sns.kdeplot(df['Item_Outlet_Sales']) # to plot the graph based on probability distribution 
```

Univariate analysis of Categorical column

```python
sns.countplot("Item_Fat_Content",data=df)
```

```python
sns.barplot(x="Item_Fat_Content",y="Item_Outlet_Sales",data=df)
```

```python
df.groupby("Item_Fat_Content").mean() 
```

Box Plot - It consists of 5 points - minimum point, q1-25%,q2-50%,q3-75%,maximum_point

```python
sns.boxplot(x="Item_Outlet_Sales",data=df)
```

```python
sns.boxplot(x="Item_Outlet_Sales",y="Item_Fat_Content",data=df)
```

Multivarient Analysis

```python
sns.relplot(x="Item_Outlet_Sales",y="Item_MRP",col="Item_Fat_Content",hue="Outlet_Type",style="Outlet_Type",
            kind="scatter",data = df,size ="Outlet_Type",sizes=(100,200))
```
